# import require modules

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import train_test_split
```

# importing dataset

```python
df=pd.read_csv("titanic_dataset(1).csv")
```

```python
df
```

# splitting the data into independent and dependent variables

```python
x=df[["PassengerId","Pclass","Sex","Age","SibSp","Parch","Fare"]]
y=df[["Survived"]]
```

# splitting the dataframe into train and test

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=450)
```

```python
test_x
```

```python
test_y
```

# Training tha model

```python
l=LogisticRegression()
l.fit(train_x,train_y)
import warnings
warnings.filterwarnings("ignore")
```

# Predicting the model

```python
l.predict([[519,2,1,36,1,0,26]])
```

```python
l.predict([[434,3,0,17,0,0,7]])
```

# Accuracy of the model

```python
l.score(train_x,train_y)
```
